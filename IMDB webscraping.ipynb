```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
pages=[]
urls=[]
title=[]
price=[]
stars=[]

scrape_pages=2
for i in range(1, scrape_pages+1):
    url=("http://books.toscrape.com/catalogue/page-{}.html").format(i)
    pages.append(url)
for item in pages:
    itm =requests.get(item)
    soup=BeautifulSoup(itm.text, "html.parser")
    #print(soup.prettify())
    for j in soup.findAll("h3"):
        tl= j.getText()
        title.append(tl)
    for p in soup.findAll("p",class_="price_color"):
        pr=p.getText()
        price.append(pr)
    for st in soup.findAll("p" , class_="star-rating"):
        for k,v in st.attrs.items():
            sta = v[1]
            stars.append(sta +" "+"stars" )
    div=soup.findAll("div", class_="image_container")
    for thumbnail in div:
        tn = thumbnail.find("img" , class_="thumbnail")
        url =("http://books.toscrape.com/" + tn['src']).replace("../" , "")
        urls.append(url)
```

```python
data={'Title' : title, 'Prices' : price, 'Rating' : stars, 'Photo' : urls}
```

```python
df = pd.DataFrame(data = data)
df.index+=1
df
```

```python
df.to_csv("scrape.csv")
```
